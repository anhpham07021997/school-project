## Relationship between price and neighboor group 

```python
price_neib=df1.groupby("neighbourhood_group")[["price","neighbourhood_group"]].mean()
price_neib
```

```python
plt.plot(price_neib.index,price_neib.values,marker="o")
plt.xticks(rotation=-45)
plt.xlabel("neighboor")
plt.ylabel("price")
```

##### => price area from west region and cetral region of Singapore are the most expensive, in contrast noth_est region is the cheapest 

# Relationship between price and neighbor

```python
price_neih=df1.groupby("neighbourhood")["price"].mean()
price_neih
```

```python

plt.figure(figsize=(50,50))

sns.barplot(price_neih.index,price_neih.values,capsize=2)
sns.plotting_context(font_scale=1.5)

plt.xticks(rotation=-45)
```

##### => region from Tuas and Southern Islands has the highest price , and Western Water Catchment is the lowest
